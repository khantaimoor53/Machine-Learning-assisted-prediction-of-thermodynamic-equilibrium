```python
#Choosing 100 unique temperature values randomly for the purpose of generating additional feed compositions/inputs 


T_points_list=[]
k=0
upper_bound = int((df_train.shape[0])*0.1) #to choose 10% of the number of datapoints in the original training dataset
for k in range(0,upper_bound):
    Temp_list=X_train_reg['Temp'].unique()
    T_gen=np.random.choice(Temp_list)
    T_points_list.append(T_gen)


```

```python
#Feed points that correspond to previously selected temperature values are extracted from the training input dataset 
#"X_train_reg" as a list and added to a list containing all feed points

#Similarily, compositions of Phases I and II corresponding to the selected Temperature values are extracted from 
#the training output dataset "y_train_reg" and added to a list containing all phase 1 and phase 2 compositions

k=0
feed_df=(X_train_reg.loc[X_train_reg['Temp'].isin(T_points_list)])
composition_df = (y_train_reg.loc[X_train_reg['Temp'].isin(T_points_list)])
all_F_points=[]
all_phase1_points=[]
all_phase2_points=[]
for i in range(0,len(T_points_list)):
    temp_selected=T_points_list[i]    
    F_point = feed_df.loc[feed_df.Temp==temp_selected,feed_df.columns[1:n+1]].values.flatten().tolist()
    all_F_points.append(F_point)
    phase1_point = composition_df.loc[feed_df.Temp==temp_selected,composition_df.columns[0:n]].values.flatten().tolist()
    all_phase1_points.append(phase1_point)
    phase2_point = composition_df.loc[feed_df.Temp==temp_selected,composition_df.columns[n:]].values.flatten().tolist()
    all_phase2_points.append(phase2_point)
```

```python
#Function to generate a point between two given points P1 and P2 such that they all lie on the same line

#uses the general parametric equation of a line from a point 

#a[0,1,2,3,4]  b[2,4,5,6,7]

#a + t*d

def gen_new_point(P1,P2,t):
    vtr1=np.array(P1)
    vtr2=np.array(P2)

    if(np.sum(vtr2)>=np.sum(vtr1)):
        d=vtr2-vtr1
        r=vtr1+t*d
    else:
        d=vtr1-vtr2
        r=vtr2+t*d
    return r.tolist()    
```

```python
#For every temperature value selected, the following code:
# 1. Finds the equation of the straight line using the extracted feed and phase coordinates
# 2. Generates 5 points between Phase 1 coordinate and Feedpoint coordinate
# 3. Generates 5 points between Phase 2 coordinate and Feedpoint coordinate
# 4. Appends all the new feed points and their corresponding phase coordinate values to respective lists

k=0
t_list=[]
for k in range(0,10):
    t= np.random.uniform(0,1)
    t_list.append(t)

F_point=[]
all_new_F_points=[]
new_F_point=[]
phase1_point = []
all_new_phase1_points =[]
phase2_point = []
all_new_phase2_points =[]
temp_augment=[]
k=0
for k in range(0,len(all_F_points)):
    
    temp_current=[T_points_list[k]]
    F_point = all_F_points[k]
    phase1_point = all_phase1_points[k]
    phase2_point = all_phase2_points[k]
    s=0
    for s in range(0,5):
        
        new_F_point=gen_new_point(phase1_point,F_point,t_list[s])
        all_new_F_points.append(new_F_point)
        all_new_phase1_points.append(phase1_point)
        all_new_phase2_points.append(phase2_point)
        temp_augment.append(temp_current)
               
        
    for s in range(5,10):
        
        new_F_point=gen_new_point(F_point,phase2_point,t_list[s])
        all_new_F_points.append(new_F_point)
        all_new_phase1_points.append(phase1_point)
        all_new_phase2_points.append(phase2_point)
        temp_augment.append(temp_current)
```

```python
#code to check if the predicted outputs of NN2 model comply with Equation 1 of the task

#For complying with equation 1:
# 1. The test feed inputs of the NN2 model and the predicted phase outputs should lie on the same line
# 2. This means that the feed, phase 1 and phase 2 coordinates must satisfy the three point collinearity constraint
# 3. Three points A,B and C are collinear if AB + BC = AC (when B lies between A and C)
# 4. The following code splits the predictions matrix into separate lists for phase 1 and phase 2
# 5. The distances between each of the three points are calculated and collinearity constraint is checked
# 6. The collinearity constraint is checked for each predicted output and a counter variable tracks if it is satisfied

k=0

dist_1_list=[]
dist_2_list=[]
dist_tot_list=[]
for k in range(0,len(collinear_input_check)):
    F_point_check=collinear_input_check[k]
    F_array=np.array(F_point_check)
    
    phases_list=collinear_output_check[k]
    i=0
    hold_phase_list=[phases_list[i:i+n] for i in range(0,len(phases_list),n)]
    
    phase1_list=hold_phase_list[0]
    phase1_array=np.array(phase1_list)
    
    phase2_list=hold_phase_list[1]
    phase2_array=np.array(phase2_list)
    
    dist_1=np.linalg.norm(F_array-phase1_array)
    dist_1_list.append(dist_1)
    dist_2=np.linalg.norm(phase2_array-F_array)
    dist_2_list.append(dist_2)
    dist_tot=np.linalg.norm(phase2_array-phase1_array)
    dist_tot_list.append(dist_tot)

L=len(dist_1_list)

s=0
counter=0
tol=0.01
for s in range(0,L):
    approx_tot=((dist_1_list[s]+dist_2_list[s]).round(2))
    if(np.abs((approx_tot-dist_tot_list[s]))<=tol):
        counter=counter+1
        
print('Equation 1 has been satisfied for %d out of %d predicted model outputs'%(counter,len(collinear_output_check)))
```
