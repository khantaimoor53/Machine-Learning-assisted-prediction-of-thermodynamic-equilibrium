```python
# NN2 training input creation - for regression

X_train_reg=df_train.drop(df_train[df_train['isSinglePhase']==1].index)
X_train_reg=X_train_reg.drop(X_train_reg.columns[n+1:],axis=1)
```

```python
# NN2 training output creation - for regression

y_train_reg=df_train.drop(df_train[df_train['isSinglePhase']==1].index)

y_train_reg=y_train_reg.drop(y_train_reg.columns[0:n+1],axis=1)
y_train_reg=y_train_reg.drop(y_train_reg.columns[-3:],axis=1)
```

```python
X_grid_reg,X_notuse,y_grid_reg,y_notuse = train_test_split(X_train_reg,y_train_reg,test_size=0.2)
```

```python
input_dim = X_grid_reg.shape[1]


def define_model(activation = 'relu',init_weights='uniform',optimizer ='Adam'):   
    model = Sequential()
    model.add(Dense(128, activation=activation, kernel_initializer=init_weights, 
                    input_dim = input_dim)) 
    model.add(Dropout(0.1))
    model.add(Dense(256, kernel_initializer=init_weights, activation=activation))
    model.add(Dense(1, kernel_initializer=init_weights, activation='sigmoid'))
    # compile the model
    model.compile(loss='mean_squared_error',optimizer=optimizer,metrics=['mae'])
    return model
```

```python
print(input_dim)
```

```python
epochs = 10

model = KerasRegressor(build_fn=define_model,epochs=epochs,verbose=1)

activation = ['tanh', 'relu', 'sigmoid']
init_weights = ['uniform', 'normal', 'he_uniform']
optimizer = ['SGD', 'RMSprop', 'Adam']
```

```python
parameter_grid = dict(activation=activation, init_weights=init_weights, optimizer=optimizer)
```

```python
print(parameter_grid)
```

```python
grid = GridSearchCV(estimator=model, param_grid=parameter_grid,cv=3)
```

```python
grid_result = grid.fit(X_grid_reg, y_grid_reg)
```

```python
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))

mean_score = grid_result.cv_results_['mean_test_score']
std_score = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(mean_score, std_score, params):
    print("Mean = %f (std=%f) with: %r" % (mean, stdev, param))
```
