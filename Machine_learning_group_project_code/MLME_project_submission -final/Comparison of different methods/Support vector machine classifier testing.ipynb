```python
# Support vector machine - training input creation for binary classification 

X_train_clas=df_train.drop(df_train.columns[n+1:],axis=1)
```

```python
print(X_train_clas) 
```

```python
#Training output creation for binary classification 

y_train_clas=df_train.drop(df_train.columns[0:3*n+3],axis=1)
```

```python
print(y_train_clas)
```

```python
#SVM Feature scaling of training input

scaler=StandardScaler()
scaler.fit(X_train_clas)

X_train_clas_scaled=scaler.transform(X_train_clas)
```

```python
print(X_train_clas_scaled)
```

```python
#Classification training using SVM

svm_model_clas = svm.SVC(kernel='rbf',C=100, tol=1e-5)
svm_model_clas.fit(X_train_clas_scaled,y_train_clas)
```

```python
y_train_clas_np = y_train_clas.to_numpy()
```

```python
predicted_train_values = (svm_model_clas.predict(X_train_clas_scaled)).reshape(-1,1)
```

```python
print(predicted_train_values)
```

```python
#Evaluating the training performance of SVM by checking the overall accuracy of training prediction,
# the number of bad outputs (real/validation output = 1 and prediction = 0) and
# the number of lost datapoints (real/validation output = 0 and prediction = 1)

a=0
b=0
bad_outputs=0
points_lost=0
no_points=X_train_clas.shape[0]
for i in range(no_points):

    if y_train_clas_np[i]==predicted_train_values[i]:
        a=a+1
    else:
        b=b+1
        if (y_train_clas_np[i]==1 and predicted_train_values[i]==0):
            bad_outputs=bad_outputs+1
        else:
            points_lost=points_lost+1
        
print('overall training accuracy of svm is',a/(a+b))
print('number of predictions that classify single phase data as multi-phase', bad_outputs)
print('number of predictions that classify multiple phase data as single phase', points_lost)
```

```python
print(X_train_clas.shape[0])
```

```python
#Training inputs for phase prediction using NN2

X_train_reg=df_train.drop(df_train[df_train['isSinglePhase']==1].index)
X_train_reg=X_train_reg.drop(X_train_reg.columns[n+1:],axis=1)


```

```python
print(X_train_reg)
```

```python
#Training outputs for phase prediction using NN2

y_train_reg=df_train.drop(df_train[df_train['isSinglePhase']==1].index)

y_train_reg=y_train_reg.drop(y_train_reg.columns[0:n+1],axis=1)
y_train_reg=y_train_reg.drop(y_train_reg.columns[-3:],axis=1)
```

```python
print(y_train_reg)
```

```python
# Standard scaling of features for NN2


scaler=StandardScaler()
print(scaler.fit(X_train_reg))
print(scaler.mean_)
X_train_reg_scaled=scaler.transform(X_train_reg)
print(X_train_reg_scaled)
```

```python
#Building NN2 for phase composition prediction


model_reg=Sequential()

#hidden layers
model_reg.add(Dense(128,input_dim=n+1,activation='relu'))
model_reg.add(Dense(256,activation='relu'))




#output layer
model_reg.add(Dense(2*n,activation='linear'))

model_reg.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=['mae'])




```

```python
print(model_reg.summary())
```

```python
#Training NN2 for predicting phase compositions

history_reg=model_reg.fit(X_train_reg_scaled,y_train_reg,validation_split=0.2,epochs=100)
```

```python
#Plotting training and validation loss of NN2

loss_reg=history_reg.history['loss']
val_loss_reg=history_reg.history['val_loss']
epochs_reg=range(1,len(loss_reg)+1)
plt.plot(epochs_reg,loss_reg,'y',label='training loss')


plt.plot(epochs_reg,val_loss_reg,'r',label='validation loss')


plt.title('Training and validation loss')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('loss') # loss is abst error
plt.legend()

plt.show()
```

```python
#Plotting training and validation dloss of NN2


acc_reg=history_reg.history['mae']


val_acc_reg=history_reg.history['val_mae']
epochs_reg=range(1,len(loss_reg)+1)
plt.plot(epochs_reg,acc_reg,'y',label='training mae')


plt.plot(epochs_reg,val_acc_reg,'r',label='validation mae')


plt.title('Training and validation mae')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('accuracy') # loss is abst error
plt.legend()

plt.show()
```
