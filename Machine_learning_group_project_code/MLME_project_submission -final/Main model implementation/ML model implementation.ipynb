```python
# Creation of training input dataframe for binary classification (NN1)

X_train_clas=df_train.drop(df_train.columns[n+1:],axis=1)
```

```python
print(X_train_clas) 
```

```python
#Creation of training output dataframe for binary classification (NN1)

y_train_clas=df_train.drop(df_train.columns[0:3*n+3],axis=1)
```

```python
print(y_train_clas)
```

```python
#Scaling of features for the training input of NN1


scaler=StandardScaler()
scaler.fit(X_train_clas)

X_train_clas_scaled=scaler.transform(X_train_clas)
```

```python
print(X_train_clas_scaled)
```

```python
#Building NN1 model for binary classification

model_clas=Sequential() #creating model with training features using Sequential class

#adding hidden layers using Dense
model_clas.add(Dense(128,input_dim=n+1,activation='tanh')) #first layer has n+1 inputs - Temperature and feed compositions
model_clas.add(Dense(256,activation='tanh'))



#output layer

model_clas.add(Dense(1,activation='sigmoid')) #the classifier model has only 1 output: whether or not the datapoint is single phase

#using log loss function as objective function
model_clas.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) 
```

```python
#summary of NN1 model features
model_clas.summary()
```

```python
#Training the binary classifier NN1 using the scaled features

history_clas=model_clas.fit(X_train_clas_scaled,y_train_clas,validation_split=0.2,epochs=300)
```

```python
#Plotting training and validation loss of NN1

loss_clas=history_clas.history['loss']
val_loss_clas=history_clas.history['val_loss']
epochs_clas=range(1,len(loss_clas)+1)
plt.plot(epochs_clas,loss_clas,'y',label='training loss')


plt.plot(epochs_clas,val_loss_clas,'r',label='validation loss')


plt.title('Training and validation loss')
plt.xlabel('epochs') # epochs is the number of times the model is trained
plt.ylabel('loss') # loss is absolute error
plt.legend()

plt.show()
```

```python
#Plotting training and validation accuracy of NN1

acc_clas=history_clas.history['accuracy']


val_acc_clas=history_clas.history['val_accuracy']
epochs_clas=range(1,len(loss_clas)+1)
plt.plot(epochs_clas,acc_clas,'y',label='training acc')


plt.plot(epochs_clas,val_acc_clas,'r',label='validation acc')


plt.title('Training and validation acc')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('accuracy') 
plt.legend()

plt.show()
```

```python
# Standard scaling of training input for NN2


scaler=StandardScaler()
print(scaler.fit(X_train_reg))
print(scaler.mean_)
X_train_reg_scaled=scaler.transform(X_train_reg)
print(X_train_reg_scaled)
```

```python
#Building NN2 for phase composition prediction using Sequential class


model_reg=Sequential()

#hidden layers
model_reg.add(Dense(128,input_dim=n+1,activation='relu')) # n+1 features
model_reg.add(Dense(256,activation='relu'))




#output layer
model_reg.add(Dense(2*n,activation='linear')) # 2*n labels corresponding to phase I and phase II compositions

#using mean squared error as objective function
model_reg.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=['mae'])




```

```python
print(model_reg.summary())
```

```python
#Training the NN2 for predicting phase compositions

history_reg=model_reg.fit(X_train_reg_scaled,y_train_reg,validation_split=0.2,epochs=100)
```

```python
#Plotting training and validation loss of NN2

loss_reg=history_reg.history['loss']
val_loss_reg=history_reg.history['val_loss']
epochs_reg=range(1,len(loss_reg)+1)
plt.plot(epochs_reg,loss_reg,'y',label='training loss')


plt.plot(epochs_reg,val_loss_reg,'r',label='validation loss')


plt.title('Training and validation loss')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('loss') # loss is abst error
plt.legend()

plt.show()
```

```python
#Plotting training and validation dloss of NN2


acc_reg=history_reg.history['mae']


val_acc_reg=history_reg.history['val_mae']
epochs_reg=range(1,len(loss_reg)+1)
plt.plot(epochs_reg,acc_reg,'y',label='training mae')


plt.plot(epochs_reg,val_acc_reg,'r',label='validation mae')


plt.title('Training and validation mae')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('accuracy') # loss is abst error
plt.legend()

plt.show()
```
